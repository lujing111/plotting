```python
from matplotlib import cm as cm


all = pd.read_excel("学生运动数据汇总.xlsx", "Sheet5")
corrmat = all.corr()
mask = np.tri(corrmat.shape[0],k=-1)
a = np.ma.array(corrmat, mask=mask) # mask out the lower triangle
fig = plt.figure()
ax1 = fig.add_subplot(111)
cmap = cm.get_cmap('jet', 10)
cmap.set_bad('w')
ax1.imshow(a, interpolation = "nearest", cmap = cmap)
ax1.grid(True)
plt.show()
```
